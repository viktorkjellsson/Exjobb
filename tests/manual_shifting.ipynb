```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
from pathlib import Path
import plotly.graph_objects as go
```

```python
# Add the root project directory to the Python path
project_root = Path.cwd().parent  # This will get the project root since the notebook is in 'notebooks/'
sys.path.append(str(project_root))

from configs.path_config import EXTRACTED_DATA_DIR

base_dir = EXTRACTED_DATA_DIR
file = 'S-B_Close_Comp.txt_IV,B,0.06.csv'
path = Path(base_dir) / 'group_alvbrodel' / file

print(path)

df = pd.read_csv(path, parse_dates=['Time'])

df = df.iloc[12216:-1]

from src.processing import remove_outliers

df = remove_outliers.clean_zeros_outliers(df)

df_orginal = df.copy()

#Shift sequence here. 
# Define indices
start_idx = 21749  # Start index of the sequence to shift
end_idx = 26337    # End index (inclusive)
reference_idx = 21620  # The point we want to align the start of the sequence to

# Calculate shift amount
shift_value = df.loc[reference_idx, 'Strain'] - df.loc[start_idx, 'Strain']

# Apply shift to the selected sequence
df.loc[start_idx:end_idx, 'Strain'] += shift_value


fig = go.Figure()

# Add traces for the different data series
fig.add_trace(go.Scatter(x=df['Time'], y=df['Strain'], mode='lines', name='Shifted data', line=dict(color='green')))
fig.add_trace(go.Scatter(x=df_orginal['Time'], y=df_orginal['Strain'], mode='lines', name='Preshift data', line=dict(color='red', dash='dash')))


# Add labels
fig.update_layout(
    title=f'Strain for {file}',
    xaxis_title='Time',
    yaxis_title='Strain',
    template='plotly_white'
)


# Show the plot
fig.show()
```

```python
"""
Code that saves the data in a csv file where they replace the df['Strain'] with df_filled['Strain'] values



"""
# Create new directory path
shifted_dir = Path(base_dir) / 'group_alvbrodel_shifted'
shifted_dir.mkdir(parents=True, exist_ok=True)  # Create the folder if it doesn't exist

# Create the new file path
shifted_path = shifted_dir / file

# Copy the original data
df_to_save = df_orginal.copy()

# Replace only the shifted segment of the 'Strain' column
df_to_save.loc[start_idx:end_idx, 'Strain'] = df.loc[start_idx:end_idx, 'Strain']

# Save the updated DataFrame
df_to_save.to_csv(shifted_path, index=False)

print(f"Partially shifted data saved to: {shifted_path}")
```
